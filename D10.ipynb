```python
#11.LIVE VIDEO FROM THE WEBCAM

import cv2
cap = cv2.VideoCapture(0) #0 considers the default webcam port

while True:
    ret,frame = cap.read() #from cap var we read the input video
    #ret and frame are 2 var to be considered, we only use frame, ret is dummy

    cv2.imshow('my live sketch', frame)
    if cv2.waitKey(1) == 13:
        break

cap.realease() #releases the webcam port
cv2.destroyAllWindows()
```

```python
#12.MY CANNY (EDGE DETECTION TECHNIQUE) SKETCH

import cv2
cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    canny = cv2.canny(frame,20,150) #20,150 are the threshold values for best edge detection
    cv2.imshow('MY CANNY SHOW', canny)

    if cv2.waitKey(1) == ord('q'):
        break

    cap.release()
    cv2.destoryAllWindows()

```

```python
#TAKE THE DATA AND CREATE DATAFRAME
import pandas as pd
df = pd.read_html("/content/DailyWiseMarketArrivals.html")[-1]
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#to remove null value
df = df.dropna()
df
```

```python
df.isnull().sum()
```

```python
#rules- the input to fb profit is always 2 cols - ds(should have relation w time) and y(price related)
#let us consider date and modal price col
#let us rename date col to ds & modal to y
df = df.rename(columns = {'Modal Price (Rs/q)' : 'y' , 'Date' : 'ds'})
df
```

```python
#let us create a new df with ds and y
df_new = df[['ds','y']]
df_new
```

```python
df_new.info()
```

```python
#the cols ds and y are not in desired datatype format 
#we convert the ds cols datatype into date time format
#we convert the y cold to integer format
df_new['ds'] = pd.to_datetime(df_new['ds'])
df_new['y'] = df_new['y'].astype(int)
df_new.info()
```

```python
#visualisatio
import matplotlib.pyplot as plt
plt.figure(figsize = (25,5))
plt.scatter(df_new['ds'], df_new['y'])
```

```python
#FB PROpHET
import prophet
op = prophet.Prophet()
op.fit(df_new)
```

```python
forecast = op.make_future_dataframe(periods=365, freq = 'D')
forecast = op.predict(forecast)
```

```python
op.plot(forecast,xlabel = 'Date', ylabel = 'potato prices/q')
#black dots - ctual potato prices
#light blue - min and max products predicted
#dark blue - best fit line
```
